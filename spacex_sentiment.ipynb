```python
from textblob import TextBlob
import pandas as pd

# Let's create some dummy tweets about SpaceX and Starlink
tweets = [
    "I absolutely love the new Starlink internet! It is incredibly fast and reliable.", # Positive
    "SpaceX rockets are okay, I guess.", # Neutral
    "My connection dropped again. Terrible service, completely useless!", # Negative
]

# Let's test the AI's "mood detector" on the first tweet
print("Testing Tweet 1: ", tweets[0])

# TextBlob reads the text
blob = TextBlob(tweets[0])

# .polarity gets the mood score from -1.0 to 1.0
score = blob.sentiment.polarity
print("\nMood Score: ", score)

# Have the AI tell us the result in plain English
if score > 0:
    print("AI says: This is POSITIVE! ")
elif score < 0:
    print("AI says: This is NEGATIVE! ")
else:
    print("AI says: This is NEUTRAL. ")
```

```python
# 1. Turn our simple list of tweets into a professional Pandas Data Table
df_tweets = pd.DataFrame({'Tweet_Text': tweets})

# 2. Build a custom "function" (a mini-machine) to calculate the mood of ANY text
def get_sentiment(text):
    score = TextBlob(text).sentiment.polarity
    if score > 0:
        return "Positive "
    elif score < 0:
        return "Negative "
    else:
        return "Neutral "

# 3. Apply our mini-machine to the entire table all at once!
df_tweets['AI_Mood_Prediction'] = df_tweets['Tweet_Text'].apply(get_sentiment)

# 4. Print the final professional table
df_tweets
```
